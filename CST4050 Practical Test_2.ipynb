#### The required libraries are imported


```python
import pandas as pd

# for scaling
from sklearn.preprocessing import StandardScaler 

# for ML Models
from sklearn.neural_network import MLPClassifier
from sklearn import svm
from sklearn.model_selection import train_test_split 
```

### Read the dataset (assuming 'data' is your data frame object)

```python
# Write the code here
filename="health_data.csv"
df=pd.read_csv(filename)
df.head()
```

### Use X,Y for the features and the target variable

```python
# Write code here
x=df.drop(["PatientID","Disease"], axis=1)
y=df["Disease"]
```


## Using the provided dataset, do the following tasks:

**Task-1: 5 points**

Standardize the features using the StandardScaler function

```python
# write code here for Task-1
scaler=StandardScaler()
x_df=scaler.fit_transform(x)
x_df=pd.DataFrame(x_df, columns=x.columns)
x_df.describe().T
```

```python
#sns.countplot(x=df["Disease"], data=df, hue="Disease")
```

**Task-2: 5 points**

Split the dataset to train and test, use 30% as test size and use 40 as the random_state

```python
# write code and discussion here for Task-2
x_train, x_test, y_train, y_test = train_test_split(
    x_df, y, test_size=0.30, random_state=40)

print("x_train shape: ",x_train.shape)
print("x_test shape: ",x_test.shape)
print("y_train shape: ",y_train.shape)
print("y_test shape: ",y_test.shape)
```

**Task-3: 10 points**

* Apply SVC to estimate the target variable. Use the following parameters: kernel='poly',  C=1
* Print the accuracy for training and testing.


```python
# write code and discussion here for Task-3
from sklearn.svm import SVC

model_svc=SVC(kernel="poly",C=1)
model_svc.fit(x_train,y_train)
```

```python
#Accuracy for training data of our SVC model
from sklearn.metrics import accuracy_score

y_pred_svc_train=model_svc.predict(x_train)

train_accuracy_svc = accuracy_score(y_train, y_pred_svc_train)
print("Train Accuracy: ",train_accuracy_svc)
```

```python
#Accuracy of test data of our SVC model

y_pred_svc_test=model_svc.predict(x_test)

test_accuracy_svc = accuracy_score(y_test, y_pred_svc_test)
print("Test Accuracy: ",test_accuracy_svc)
```

**Task-4: 15 points**

* Apply MLP to estimate the target variable. Use the following parameters: hidden_layer_sizes=(10), activation='logistic', solver='sgd',
 learning_rate='constant', learning_rate_init=0.001, max_iter=500
* Print the accuracy for training and testing.

```python
# write code and discussion here for Task-4
model_mlp = MLPClassifier(
    hidden_layer_sizes=(10,),
    activation='logistic',
    solver='sgd',
    learning_rate='constant',
    learning_rate_init=0.001,
    max_iter=500,
    random_state=40
)

model_mlp.fit(x_train, y_train)
```

```python
#Accuracy for training data of our MLP model

y_pred_mlp_train=model_mlp.predict(x_train)

train_accuracy_mlp = accuracy_score(y_train, y_pred_mlp_train)
print("Train Accuracy: ",train_accuracy_mlp)
```

```python
#Accuracy of test data of our MLP model

y_pred_mlp_test=model_mlp.predict(x_test)

test_accuracy_mlp = accuracy_score(y_test, y_pred_mlp_test)
print("Test Accuracy: ",test_accuracy_mlp)
```
